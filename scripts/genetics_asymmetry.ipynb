```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Initialize plotting

```python
matplotlib.rcParams["svg.fonttype"] = "none"
plt.rcParams.update({"font.family": "Helvetica Neue"})
cm = 1 / 2.54  # centimeters in inches

color = ['#11a579', '#f2b701', '#e73f74', '#3969ac',
         '#e68310', '#008695', '#cf1c90', '#f97b72',
         '#a5aa99', '#ba805a', '#80ba5a']
colors = []
for c in color:
    colors.append(matplotlib.colors.to_rgb(c))
```

Define functions

```python
def change_width(ax, new_value):
    for patch in ax.patches:
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
```

Load data

```python
# %% SNP association
df_snp = pd.read_csv('genes.csv', index_col=False)

df_snp = df_snp.sort_values(by=['cat', 'log(p)'], ascending=[True, False])
df_snp["phen"] = np.array(['Vertex-wise sulcal depth',
                           'Left–right brain asymmetry',
                           'Brain morphology (MOSTest)',
                           'Whole brain restricted directional\ndiffusion (multivariate analysis)',
                           'Vertex-wise cortical surface area',
                           'Cortical surface area (MOSTest)',
                           'Cortical surface area',
                           'Vertex-wise cortical thickness',
                           'Cortical thickness',
                           'Inter-alpha-trypsin inhibitor\nheavy chain H5 levels (ITIH5.8233.2.3)',
                           'Planum temporale asymmetry index',
                           'IDP T1 FAST ROIs L planum temporale',
                           'Left planum temporale\ngrey matter volume'],
                          dtype=object)
df_snp.head()
```

GWAS associations with rs41298373 based on GWAS Catalog.

```python
sns.set_theme(style="ticks", palette="colorblind", font_scale=1.8,
              font='Helvetica Neue', rc={"axes.spines.right": False,
                                         "axes.spines.top": False,
                                         "axes.spines.left": False})
fig, ax = plt.subplots(1, figsize=(len(df_snp), 6))
g = sns.barplot(x='phen', y='log(p)', hue='cat', data=df_snp,
                palette=colors)
g.set_title('')
g.set_xticklabels(g.get_xticklabels(), rotation=45,
                  horizontalalignment='right')
g.set_ylabel('-log10 P-value')
g.set_xlabel('Phenotype')
ax.yaxis.grid()
sns.despine(left=True)
change_width(ax, 0.5)
plt.show()
```
